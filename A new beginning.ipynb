```python
# connect to SQL server

starttime = time.time()

connection = connect(
            host="localhost",
            user="root",
            password=getpass("Enter password: "),
            database="bikes",
        )

while True:
    try:
        cursor = connection.cursor()
        size_str = str(getsize())
        # size_str = str(size)
        print(size_str)
        # print(size)
        now = time.strftime('%Y-%m-%d %H:%M:%S')
        cursor.execute("INSERT INTO count (count, time) VALUES (" + size_str + ", '"+ now +"')")
        connection.commit()
    except Error as e:
        print(e)

    time.sleep(60.0 - ((time.time() - starttime) % 60.0))

```
