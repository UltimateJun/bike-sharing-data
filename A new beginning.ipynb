Let us get some juicy bike data
    

```python
import requests
from requests.structures import CaseInsensitiveDict
from pprint import pprint
import json
from getpass import getpass
from mysql.connector import connect, Error
import time
```

```python
def getsize():

    lat = "&lat=52.518611"
    lon = "&lon=13.408333"
    radius = "&radius=10000"
    limit = "&limit=100"
    providernetwork = ""
    url = "https://api.deutschebahn.com/flinkster-api-ng/v1/bookingproposals?providernetwork=2" + lat + lon + radius + limit

    headers = {
        'Accept': 'application/json',
        'Authorization': 'Bearer 56b6c4f18d92c4869078102e978ec8b9',
    }

    resp = requests.get(url, headers=headers)

    size = resp.json()['size']
    print(size)
    return size
```

```python
headers = {
    'Accept': 'application/json',
    'Authorization': 'Bearer 56b6c4f18d92c4869078102e978ec8b9',
}
resp = requests.get('https://api.deutschebahn.com/flinkster-api-ng/v1/bookingproposals?lat=52.518611&lon=13.408333&radius=10000&limit=100&providernetwork=2', headers=headers)
print(json.dumps(resp.json(),indent=5))
```

```python
# connect to SQL server

starttime = time.time()

connection = connect(
            host="localhost",
            user="root",
            password=getpass("Enter password: "),
            database="bikes",
        )

while True:
    try:
        cursor = connection.cursor()
        size_str = str(getsize())
        # size_str = str(size)
        print(size_str)
        # print(size)
        now = time.strftime('%Y-%m-%d %H:%M:%S')
        cursor.execute("INSERT INTO count (count, time) VALUES (" + size_str + ", '"+ now +"')")
        connection.commit()
    except Error as e:
        print(e)

    time.sleep(60.0 - ((time.time() - starttime) % 60.0))

```
